```python
# Simon's Algorithem

# importing Qiskit
#from qiskit import IBMQ, Aer
from qiskit_aer import Aer
#from qiskit.providers.ibmq import least_busy
from qiskit import QuantumCircuit, transpile

# import basic plot tools
from qiskit.visualization import plot_histogram
from qiskit_textbook.tools import simon_oracle

# Step 0: starting state
b = '000'

# Step 1: initialise the quantum circuit, len(b) determines the number of quibits
n = len(b)
simon_circuit = QuantumCircuit(n*2, n)

# Apply barrier for visual separation
simon_circuit.barrier()

# Step 2: Hadamard * n, hadamard on each qubit with range(n)
simon_circuit.h(range(n))    
    
# Apply barrier for visual separation
simon_circuit.barrier()

# Step 3: Apply oracle query 
simon_circuit = simon_circuit.compose(simon_oracle(b))

# Apply barrier for visual separation
simon_circuit.barrier()

# Step 4: Hadamard to the qubits
simon_circuit.h(range(n))

# Measure qubits
simon_circuit.measure(range(n), range(n))
simon_circuit.draw()

# visualize
aer_sim = Aer.get_backend('aer_simulator')
results = aer_sim.run(simon_circuit).result()
counts = results.get_counts()
plot_histogram(counts)


'''
for the function used in the rest of the question 7

# define function f(x, b)
def f(x, b):
    if x == 0 or x == 7:  # x == 000 or x == 111
        return '000'
    elif x == 1 or x == 6:  # x == 001 or x == 110
        return '001'
    elif x == 2 or x == 5:  # x == 010 or x == 101
        return '010'
    elif x == 3 or x == 4:  # x == 011 or x == 100
        return '011'

# Step 3: Apply oracle query 
for q in range(n):
    simon_circuit.cx(q, n+q)  # Apply CNOT gate
    for i, b_i in enumerate(b):
        if b_i == '1':
            simon_circuit.cx(q, n+i)  # Apply CNOT gate'''
```
